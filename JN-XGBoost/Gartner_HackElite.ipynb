```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(X,Y,random_state=42)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=7)
```

```python
X_train=X_train.drop(columns=['Client ID','Company ID','Flag 4','Flag 6','Flag 3', 'Flag 1','Client Contract Starting Month'])
X_test_X=X_test.drop(columns=['Client ID','Company ID','Flag 4','Flag 6','Flag 3', 'Flag 1','Client Contract Starting Month'])
```

```python
model = xgb.XGBClassifier(max_depth=4,
                        subsample=0.9,
                        objective='binary:logistic',
                        n_estimators=400,
                        learning_rate = 0.1)
```

```python
train_X=train_X.drop(columns=['Client ID','Company ID','Flag 4','Flag 6','Flag 3', 'Flag 1','Client Contract Starting Month'])
test_X=test_X.drop(columns=['Client ID','Company ID','Flag 4','Flag 6','Flag 3', 'Flag 1','Client Contract Starting Month'])
```

```python
X.head()
```

```python
X_test=X_test.drop(columns=['Client ID','Company ID','Flag 4','Flag 6','Flag 3', 'Flag 1','Client Contract Starting Month'])
```

```python
eval_set = [(train_X, train_Y), (test_X, test_Y)]
model.fit(train_X, train_Y.values.ravel(), early_stopping_rounds=10, eval_metric=["error", "logloss"], eval_set=eval_set, verbose=True)
```

```python
d_train = xgb.DMatrix(X_train, label=y_train)
d_test = xgb.DMatrix(X_test, label=y_test)
params = {
    "eta": 0.01,
    "objective": "binary:logistic",
    "subsample": 0.5,
    "base_score": 0.5,
    "eval_metric": "logloss"
}
model2 = xgb.train(params, d_train, 5000, evals = [(d_test, "test")], verbose_eval=100, early_stopping_rounds=20)
```

```python
y_pred=model.predict(test_X)
y_pred_pd=pd.DataFrame({'Y_PRED':y_pred})
print(y_pred.size)
print(np.count_nonzero(y_pred_pd==1))
print(np.count_nonzero(test_Y==1))
y_pred
```
